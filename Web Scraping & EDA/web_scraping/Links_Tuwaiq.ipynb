```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time

# List of URLs to scrape
urls = ['https://tuwaiq.edu.sa/bootcamp/KV4waj2Z/view',
 'https://tuwaiq.edu.sa/bootcamp/4G6ZLmVJ/view',
 'https://tuwaiq.edu.sa/bootcamp/VjJ4mYlE/view',
 'https://tuwaiq.edu.sa/bootcamp/qRxVg18X/view',
 'https://tuwaiq.edu.sa/bootcamp/vleGDNVR/view',
 'https://tuwaiq.edu.sa/bootcamp/Vo05XADm/view',
 'https://tuwaiq.edu.sa/bootcamp/mog23GNL/view',
 'https://tuwaiq.edu.sa/bootcamp/xwYwPqLa/view',
 'https://tuwaiq.edu.sa/bootcamp/Mq2yzpl5/view',
 'https://tuwaiq.edu.sa/bootcamp/KwVlDV3b/view',
 'https://tuwaiq.edu.sa/bootcamp/E83oNvJ6/view',
 'https://tuwaiq.edu.sa/bootcamp/OrzZL0EZ/view',
 'https://tuwaiq.edu.sa/bootcamp/5ejGzPmd/view',
 'https://tuwaiq.edu.sa/bootcamp/P5maknX4/view',
 'https://tuwaiq.edu.sa/bootcamp/1JeYqe5k/view',
 'https://tuwaiq.edu.sa/bootcamp/wYd8Ddq4/view',
 'https://tuwaiq.edu.sa/bootcamp/xGPXRKwV/view',
 'https://tuwaiq.edu.sa/bootcamp/J0DYZmAN/view',
 'https://tuwaiq.edu.sa/bootcamp/7kp4GAB5/view',
 'https://tuwaiq.edu.sa/bootcamp/jdKDMLZl/view',
 'https://tuwaiq.edu.sa/bootcamp/Z1l5AlN3/view',
 'https://tuwaiq.edu.sa/bootcamp/QNLPN8R3/view',
 'https://tuwaiq.edu.sa/bootcamp/3oKm1KZE/view',
 'https://tuwaiq.edu.sa/bootcamp/2RPAOKZv/view',
 'https://tuwaiq.edu.sa/bootcamp/J6VQqVeA/view',
 'https://tuwaiq.edu.sa/bootcamp/mnzywexo/view',
 'https://tuwaiq.edu.sa/bootcamp/L20dYnXK/view',
 'https://tuwaiq.edu.sa/bootcamp/EA0pD0J1/view',
 'https://tuwaiq.edu.sa/bootcamp/AXgMaglM/view',
 'https://tuwaiq.edu.sa/bootcamp/qe109Am5/view',
 'https://tuwaiq.edu.sa/bootcamp/4PpxrDG2/view',
 'https://tuwaiq.edu.sa/bootcamp/E7D8wO8L/view',
 'https://tuwaiq.edu.sa/bootcamp/YZaBO63N/view',
 'https://tuwaiq.edu.sa/bootcamp/gb2DAepL/view',
 'https://tuwaiq.edu.sa/bootcamp/anVM346G/view',
 'https://tuwaiq.edu.sa/bootcamp/k26XzBP4/view',
 'https://tuwaiq.edu.sa/bootcamp/eoAw2Y16/view',
 'https://tuwaiq.edu.sa/bootcamp/K9EMrojl/view',
 'https://tuwaiq.edu.sa/bootcamp/6RYoanm7/view',
 'https://tuwaiq.edu.sa/bootcamp/JdVRpljR/view',
 'https://tuwaiq.edu.sa/bootcamp/AO2Nn2Xq/view',
 'https://tuwaiq.edu.sa/bootcamp/Z2YB2YGR/view',
 'https://tuwaiq.edu.sa/bootcamp/AjBAbxz2/view',
 'https://tuwaiq.edu.sa/bootcamp/a9Qp6rZb/view',
 'https://tuwaiq.edu.sa/bootcamp/RDxd9l7q/view',
 'https://tuwaiq.edu.sa/bootcamp/GoYpPYlK/view',
 'https://tuwaiq.edu.sa/bootcamp/ZVPgKPop/view',
 'https://tuwaiq.edu.sa/bootcamp/8vDAXa2Z/view',
 'https://tuwaiq.edu.sa/bootcamp/892YZQ3j/view',
 'https://tuwaiq.edu.sa/bootcamp/XqAz7rJM/view',
 'https://tuwaiq.edu.sa/bootcamp/nmY8Kp6g/view',
 'https://tuwaiq.edu.sa/bootcamp/zq0lRZ4V/view',
 'https://tuwaiq.edu.sa/bootcamp/aW6x5y28/view',
 'https://tuwaiq.edu.sa/bootcamp/REaV2aWq/view',
 'https://tuwaiq.edu.sa/bootcamp/2nAlPALy/view',
 'https://tuwaiq.edu.sa/bootcamp/vkmGwmle/view',
 'https://tuwaiq.edu.sa/bootcamp/qNAd6vnz/view',
 'https://tuwaiq.edu.sa/bootcamp/3bnoygLz/view',
 'https://tuwaiq.edu.sa/bootcamp/KmaBDNYp/view',
 'https://tuwaiq.edu.sa/bootcamp/GnALer8D/view',
 'https://tuwaiq.edu.sa/bootcamp/qYJdbmaR/view',
 'https://tuwaiq.edu.sa/bootcamp/J4Q46rqK/view',
 'https://tuwaiq.edu.sa/bootcamp/XpZV75VY/view',
 'https://tuwaiq.edu.sa/bootcamp/pLAO6K3w/view',
 'https://tuwaiq.edu.sa/bootcamp/bKNaj1yx/view',
 'https://tuwaiq.edu.sa/bootcamp/l0PB01qJ/view',
 'https://tuwaiq.edu.sa/bootcamp/3Rr1KQ8J/view',
 'https://tuwaiq.edu.sa/bootcamp/YkxW3xr9/view',
 'https://tuwaiq.edu.sa/bootcamp/2n4ODWmQ/view',
 'https://tuwaiq.edu.sa/bootcamp/lMrDe24g/view',
 'https://tuwaiq.edu.sa/bootcamp/x2m7Ny3n/view',
 'https://tuwaiq.edu.sa/bootcamp/mj5d969Q/view',
 'https://tuwaiq.edu.sa/bootcamp/YGQ9ZARZ/view',
 'https://tuwaiq.edu.sa/bootcamp/6grZK2nj/view',
 'https://tuwaiq.edu.sa/bootcamp/l63b9pW8/view',
 'https://tuwaiq.edu.sa/bootcamp/qmY8Er0K/view',
 'https://tuwaiq.edu.sa/bootcamp/6Zm3wmbK/view',
 'https://tuwaiq.edu.sa/bootcamp/gRon7rd4/view',
 'https://tuwaiq.edu.sa/bootcamp/WVRdmPEv/view',
 'https://tuwaiq.edu.sa/bootcamp/1gp2Wakd/view',
 'https://tuwaiq.edu.sa/bootcamp/qjAoBQV6/view',
 'https://tuwaiq.edu.sa/bootcamp/exX4KYn2/view',
 'https://tuwaiq.edu.sa/bootcamp/Dvl7VOzE/view',
 'https://tuwaiq.edu.sa/bootcamp/zkO0eN0l/view',
 'https://tuwaiq.edu.sa/bootcamp/972JDLvV/view',
 'https://tuwaiq.edu.sa/bootcamp/vX3Xp3ol/view',
 'https://tuwaiq.edu.sa/bootcamp/q2lZ8VpW/view',
 'https://tuwaiq.edu.sa/bootcamp/gEzLm9WV/view',
 'https://tuwaiq.edu.sa/bootcamp/6qaMoNMG/view',
 'https://tuwaiq.edu.sa/bootcamp/6zJngQWV/view',
 'https://tuwaiq.edu.sa/bootcamp/ApeVneKD/view',
 'https://tuwaiq.edu.sa/bootcamp/7yaxjXvn/view',
 'https://tuwaiq.edu.sa/bootcamp/6m0O70r5/view',
 'https://tuwaiq.edu.sa/bootcamp/kR9DMO8z/view',
 'https://tuwaiq.edu.sa/bootcamp/v8AwQbKW/view',
 'https://tuwaiq.edu.sa/bootcamp/Ar6V0xK9/view',
 'https://tuwaiq.edu.sa/bootcamp/vQpOb2q9/view',
 'https://tuwaiq.edu.sa/bootcamp/8zrJW2Kq/view',
 'https://tuwaiq.edu.sa/bootcamp/kv1axgEn/view',
 'https://tuwaiq.edu.sa/bootcamp/Xm07YKod/view',
 'https://tuwaiq.edu.sa/bootcamp/8yV5X2gp/view',
 'https://tuwaiq.edu.sa/bootcamp/MD2RN2Le/view',
 'https://tuwaiq.edu.sa/bootcamp/PDJYkq7n/view',
 'https://tuwaiq.edu.sa/bootcamp/Rg4nL36r/view',
 'https://tuwaiq.edu.sa/bootcamp/a0YmeY82/view',
 'https://tuwaiq.edu.sa/bootcamp/pmxeXBOd/view',
 'https://tuwaiq.edu.sa/bootcamp/0dGbx2x6/view',
 'https://tuwaiq.edu.sa/bootcamp/Ba2lyQ64/view',
 'https://tuwaiq.edu.sa/bootcamp/kXA76Qb6/view',
 'https://tuwaiq.edu.sa/bootcamp/MZjpEBz0/view',
 'https://tuwaiq.edu.sa/bootcamp/eq3WNEBj/view',
 'https://tuwaiq.edu.sa/bootcamp/bz2ZWRkL/view',
 'https://tuwaiq.edu.sa/bootcamp/rV5qbmDd/view',
 'https://tuwaiq.edu.sa/bootcamp/XRJwQnK2/view',
 'https://tuwaiq.edu.sa/bootcamp/v1AmlQjw/view',
 'https://tuwaiq.edu.sa/bootcamp/aE7xwOLe/view',
 'https://tuwaiq.edu.sa/bootcamp/GmowzA4p/view',
 'https://tuwaiq.edu.sa/bootcamp/X4ExoPzo/view',
 'https://tuwaiq.edu.sa/bootcamp/VgMYQMd1/view',
 'https://tuwaiq.edu.sa/bootcamp/wvp8Xpz1/view',
 'https://tuwaiq.edu.sa/bootcamp/EGWB7V61/view',
 'https://tuwaiq.edu.sa/bootcamp/MDQGYrG8/view',
 'https://tuwaiq.edu.sa/bootcamp/pB7WYO4V/view',
 'https://tuwaiq.edu.sa/bootcamp/NX4zb4V8/view',
 'https://tuwaiq.edu.sa/bootcamp/JEYKLrnL/view',
 'https://tuwaiq.edu.sa/bootcamp/DEL50Pog/view',
 'https://tuwaiq.edu.sa/bootcamp/kO2yjQlx/view',
 'https://tuwaiq.edu.sa/bootcamp/lBzJpG0w/view',
 'https://tuwaiq.edu.sa/bootcamp/oQR7ORP5/view',
 'https://tuwaiq.edu.sa/bootcamp/E0QXqxPO/view',
 'https://tuwaiq.edu.sa/bootcamp/gvrw5QV3/view',
 'https://tuwaiq.edu.sa/bootcamp/vQzNnz7Z/view',
 'https://tuwaiq.edu.sa/bootcamp/73DyMDg5/view',
 'https://tuwaiq.edu.sa/bootcamp/M6DKQP13/view',
 'https://tuwaiq.edu.sa/bootcamp/3aLGN0KR/view',
 'https://tuwaiq.edu.sa/bootcamp/kyGPqnYz/view',
 'https://tuwaiq.edu.sa/bootcamp/9VzGa6gJ/view',
 'https://tuwaiq.edu.sa/bootcamp/WlPoMRME/view',
 'https://tuwaiq.edu.sa/bootcamp/6xqbPJkZ/view',
 'https://tuwaiq.edu.sa/bootcamp/JKgoQ5vR/view',
 'https://tuwaiq.edu.sa/bootcamp/45GerpZz/view',
 'https://tuwaiq.edu.sa/bootcamp/kpxremZV/view',
 'https://tuwaiq.edu.sa/bootcamp/jX6WNYAk/view',
 'https://tuwaiq.edu.sa/bootcamp/4bpq42r6/view']

# Initialize the webdriver
driver = webdriver.Chrome()

# Initialize lists to store the extracted information
Course_names = []
descriptions = []
goals = []
locations = []
date_ranges = []
durations = []
start_dates = []
end_dates = []
Age_groups = []
questions_list = []
answers_list = []

for url in urls:
    # Open the URL
    driver.get(url)
    time.sleep(3)  # Wait for the page to load

    # Parse the page content
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Extract the course title
    title = soup.find('h1', class_='text-2xl md:text-3xl font-bold text-[#101828] break-words text-center').text
    Course_names.append(title)

    # Extract the description
    description = soup.find('p', class_='break-words').text
    descriptions.append(description)

    # Extract the goals
    ul_element = soup.find('ul', class_='list-disc')
    li_elements = ul_element.find_all('li')
    goals_list = [li.text for li in li_elements]
    goals.append(goals_list)

    # Extract the location information
    location_element = soup.select_one('span[x-text="course?.locationText"]')
    if location_element:
        locations.append(location_element.text)

    # Extract the date range information
    date_range_element = soup.select_one('h6[x-text^="humanizeDateRange"]')
    if date_range_element:
        date_ranges.append(date_range_element.text)

    # Extract the duration information
    duration_element = soup.select_one('h6[x-text="course?.durationText"]')
    if duration_element:
        durations.append(duration_element.text)

    # Extract the start and end dates
    start_date = soup.select_one('div:nth-of-type(1) > h6.break-words').text
    end_date = soup.select_one('div:nth-of-type(2) > h6.break-words').text
    start_dates.append(start_date)
    end_dates.append(end_date)

    # Extract the FAQ section
    faq_div = soup.find('div', class_='p-3 px-11')
    faq_items = faq_div.find_all('li', class_='!mt-5')

    # Initialize lists to store the questions and answers
    questions = []
    answers = []

    # Loop through each FAQ item and extract the question and answer
    for item in faq_items:
        question = item.find('h4').text
        answer = item.find('p').text
        questions.append(question)
        answers.append(answer)

    questions_list.append(questions)
    answers_list.append(answers)

    # Extract the age group information
    age = soup.find('span', class_='text-xl').text
    Age_groups.append(age)

# Close the driver
driver.quit()

# Print the extracted data
print("Course Names:", Course_names)
print("Descriptions:", descriptions)
print("Goals:", goals)
print("Locations:", locations)
print("Date Ranges:", date_ranges)
print("Durations:", durations)
print("Start Dates:", start_dates)
print("End Dates:", end_dates)
print("Age Groups:", Age_groups)
print("Questions:", questions_list)
print("Answers:", answers_list)
```
