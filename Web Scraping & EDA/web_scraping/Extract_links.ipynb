```python
initiative_category_names = []

# Find all <p> elements
category_elements = soup.find_all('p')

# Define the specific class patterns to match
target_classes = [
    'w-full text-lg text-[#C084FC]',
    'w-full text-lg text-[#91D8DB]',
    'w-full text-lg text-[#ABADF5]',
    'w-full text-lg text-[#B4B4B4]',
    'w-full text-lg text-[#BEE5C2]'
]

# Check each element and append to the list if it meets the criteria
for element in category_elements:
    # Get the value of the class attribute
    class_attr = ' '.join(element.get('class', []))
    
    # Check if the class attribute matches any of the target classes
    if class_attr in target_classes:
        # Extract the inner text
        initiative_category_names.append(element.text.strip())

print(initiative_category_names)
```

```python
#scrolling 
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

# Initialize the WebDriver (make sure to replace 'chromedriver' with the path to your WebDriver)
driver = webdriver.Chrome()

# Open the Tuwaiq Academy bootcamps page
driver.get('https://tuwaiq.edu.sa/bootcamps?type=NORMAL')
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'html.parser')
# Initialize the WebDriver (make sure to replace 'chromedriver' with the path to your WebDriver)
driver = webdriver.Chrome()

# Open the Tuwaiq Academy bootcamps page
driver.get('https://tuwaiq.edu.sa/bootcamps?type=NORMAL')
# Wait for the page to load
time.sleep(5)

# Scroll the page to ensure all elements are loaded
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(2)  # Adjust sleep time if needed

# Define a function to click the "Load More" button
def click_load_more():
    try:
        # Use XPath to locate the "Load More" button based on its class
        load_more_button = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, '//button[contains(@class, "rounded-full") and contains(@class, "border") and contains(@class, "transition-colors") and contains(@class, "duration-300") and contains(@class, "p-2") and contains(@class, "px-8") and contains(@class, "mx-auto") and contains(@class, "text-[#4E3797]") and contains(@class, "hover:text-white") and contains(@class, "border-[#4E3797]") and contains(@class, "hover:bg-[#4E3797]") and contains(@class, "bg-white")]')
        ))
        print("Scrolling to 'Load More' button.")
        driver.execute_script("arguments[0].scrollIntoView(true);", load_more_button)
        time.sleep(2)  # Wait for the button to be fully in view

        print("Clicking 'Load More' button.")
        driver.execute_script("arguments[0].click();", load_more_button)
        return True
    except Exception as e:
        print(f"Exception occurred: {e}")
        return False

# Keep clicking "Load More" until all data is loaded
while click_load_more():
    time.sleep(5)  # Adjust the sleep time as needed

# Get the page source and parse it with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Extract the desired data
# Adjust the selector based on the actual website structure
bootcamps = soup.find_all('div', class_='bootcamp-card')  # Update this with the correct class name

# Loop through the extracted data and print it
for bootcamp in bootcamps:
    title = bootcamp.find('h3').text.strip()  # Update this with the correct tag and class
    description = bootcamp.find('p').text.strip()  # Update this with the correct tag and class
    print(f'Title: {title}')
    print(f'Description: {description}')
    print('---')

# Close the WebDriver
driver.quit()
```
