```python
from weaviate.util import get_valid_uuid
from uuid import uuid4
from tqdm import tqdm
import time
```

```python
start = time.time()

# Inspect the DataFrame columns
print(df_courses.columns)

# upload courses
for idx, row in tqdm(df_courses.iterrows(), total=df_courses.shape[0]):
    data_properties = {
        "course_name": row["course_name"],  # Use the correct column name
        "course_description": row["course_description"],
        "learning_goals": row["learning_goals"],
        "level": row["level"],
        "duration": row["duration"],
        "exam": row["exam"],
        "project": row["project"],
        "course_content": row["course_content"],
        "skills": row["skills"],
        "content": row["tools_languages"],
        "job_titles": row["job_titles"]
    }
    # id = get_valid_uuid(uuid4())
    with client.batch as batch:
        batch.add_data_object(
            data_properties,
            "Courses"
        )

print(f"Uploaded {idx + 1} documents in {time.time() - start} seconds.")
```

```python
# Inspect the DataFrame columns
print(df_TQ.columns)


# Upload data
start = time.time()

for idx, row in tqdm(df_TQ.iterrows(), total=df_TQ.shape[0]):
    data_properties = {
        "content": row["course_name"],
        "typeOfProgram": row["type_of_program"],
        "description": row["description"],
        "goals": row["goals"],
        "location": row["location"],
        "date_range": row["date_range"],
        "duration": row["duration"],
        "start_date": row["start_date"],
        "end_date": row["end_date"],
        "age_group": row["age_group"],
        "questions": row["questions"],
        "answers": row["answers"],
        "links": row["links"],
        "tools_and_languages": row["tools_and_languages"],
        "acquired_skills": row["acquired_skills"],
        "job_titles": row["job_titles"]
    }

    with client.batch as batch:
        batch.add_data_object(
            data_properties,
            "TQSkills"
        )

print(f"Uploaded {idx + 1} documents in {time.time() - start} seconds.")
```
